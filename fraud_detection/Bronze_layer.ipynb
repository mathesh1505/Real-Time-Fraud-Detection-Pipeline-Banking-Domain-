```python
%sql
CREATE VOLUME fraud_detection.bronze.chk_volume;
```

```python

spark.sql("USE CATALOG fraud_detection")
spark.sql("USE SCHEMA bronze")

df = spark.readStream.table("fraud_detection.source.transactions")

df.writeStream \
  .format("delta") \
  .option("checkpointLocation", "/Volumes/fraud_detection/bronze/chk_volume/transactions_chk") \
  .trigger(availableNow=True) \
  .toTable("transactions_bronze")
```

```python
%sql
SELECT count(*) FROM fraud_detection.bronze.transactions_bronze;
```
